```python
#@title Texto de título padrão
a=int(input('escreva um valor para a:'))
b=int(input('escreva um valor para b:'))

print('digite 1 para somar')
print('digite 2 para multiplicar')
print('digite 3 para subtrair')
print('digite 4 para dividir')
print('digite 5 para somar os quadrados')
print('digite 6 para spmar os cubos')

c=int(input('o que você deseja calcular?'))

if c==1:
 print('o resultado da soma é:',a+b)

elif c==2:
  print('o resultado da multiplicação é:',a*b)

elif c==3:
  print('o resultado da subtração a-b:',a-b)
  print('o resultado da subtração b-a:',b-a)

elif c==4:
  print('resultado da divisão a/b:',a/b)
  print('resultado do resto da divisão a/b:',a//b)
  print('resultado da divisão b/a:',b/a)
  print('resultado do resto da divisão b/a:',b//a)

elif c==5:
  print('resultado do quadrado a:',a**2)
  print('resultado do quadrado b:',b**2)

elif c==6:
  print('resultado do cubo a:',a**3)
  print('resultado do cubo b:',b**3)

else:
  print('valor invalido')
```
